```python
import pandas as pd
import numpy as np
import seaborn as sns
from plot_helper import remove_rank, calculate_wall_clock_time_thresholds, calculate_lengths_achieved, create_population_barplot, create_scaling_lineplot
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
data_frames = {
    "a280"  : "population_test_a280_final.gz",
    "d1291" : "population_test_d1291_final.gz",
    "u2319" : "population_test_u2319_final.gz",
    "bier127" : "population_test_bier127_final.gz"
}

# Color map for population plots
cmap = "plasma"
```

# a280

```python
df = pd.read_csv(data_frames["a280"])
thresholds = calculate_wall_clock_time_thresholds(df)
line_df = calculate_lengths_achieved(df, thresholds)
create_population_barplot(line_df, "a280", thresholds, palette=cmap)
```
