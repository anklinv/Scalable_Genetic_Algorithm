```python
import seaborn as sns
import pandas as pd
from plot_helper import remove_rank, calculate_wall_clock_time_thresholds, calculate_lengths_achieved, create_scaling_lineplot
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

# Parameters

```python
data_frames = {
    "a280"  : "scaling_test_a280_final.gz",
    "d1291" : "scaling_test_d1291_final.gz",
    "u2319" : "scaling_test_u2319_final.gz",
    "bier127" : "scaling_test_bier127_Jan_10_113307.gz"
}

y_scaling_limits = {
    "a280" : [
        3500,
        5000
    ],
    "d1291" : [
        110000,
        400000
    ],
    "u2319" : [
        380000,
        1500000
    ],
    "bier127" : [
        120000,
        135000
    ]
}

# Defines which lines to show in the legend of the scaling plot
legend_n = [1, 2, 4, 8, 12, 16, 24, 32, 48, 64]

# Legend in two columns
legend_col = 2

# Color map for scaling plots
cmap = "plasma"
```

# bier127

```python
df = pd.read_csv(data_frames["bier127"])
df = remove_rank(df)
thresholds = calculate_wall_clock_time_thresholds(df)
line_df = calculate_lengths_achieved(df, thresholds)
create_scaling_lineplot(line_df, "bier127", palette=cmap,
                        manual_legend=True, legend_n=legend_n, # manual legend override
                        limits=True, y_scaling_limits=y_scaling_limits, legend_col=legend_col) # manual limits override
```

# a280

```python
df = pd.read_csv(data_frames["a280"])
df = remove_rank(df)
thresholds = calculate_wall_clock_time_thresholds(df)
line_df = calculate_lengths_achieved(df, thresholds)
create_scaling_lineplot(line_df, "a280", palette=cmap,
                        manual_legend=True, legend_n=legend_n, # manual legend override
                        limits=True, y_scaling_limits=y_scaling_limits, legend_col=legend_col) # manual limits override
create_scaling_lineplot(line_df, "a280", palette=cmap, limits=False,
                       manual_legend=True, legend_n=legend_n, legend_col=legend_col)  # manual legend override
```
