```python
import seaborn as sns
import pandas as pd
from plot_helper import remove_rank, create_scaling_speedup_plot
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
data_frames = {
    "a280"  : "scaling_test_a280_final.gz",
    "d1291" : "scaling_test_d1291_final.gz",
    "u2319" : "scaling_test_u2319_final.gz"
}

# Defines which limit to use for the speedup plots
y_scaling_limits_speedup = {
    "a280"  : 40,
    "d1291" : 20,
    "u2319" : 14
}

# Defines which ticks to use for the speedup
yticks_speedup = {
    "a280"  : [1, 10, 20, 30, 40],
    "d1291" : [1, 4, 8, 12, 16, 20],
    "u2319" : [1, 4, 7, 10, 13]
}

# Defines which thresholds to use for the speedup plots
scaling_thresholds = {
    "a280" : [
        3800,
        3900,
        4000,
        4100,
        4200,
        4300  
    ],
    "d1291" : [
        180000,
        200000,
        300000
    ],
    "u2319" : [
        873600,
        900000,
        1000000,
        1200000,
        1400000,
        1800000
    ]
}

# Ticks for speedup plot
xticks_n = [1, 4, 8, 16, 24, 32, 48, 64]
```

# a280

```python
df = pd.read_csv(data_frames["a280"])
df = remove_rank(df)
create_scaling_speedup_plot(df, "a280", xticks_n=xticks_n, scaling_thresholds=scaling_thresholds,
                            static_ylim=True, y_scaling_limits_speedup=y_scaling_limits_speedup, yticks_speedup=yticks_speedup)
```

# d1291

```python
df = pd.read_csv(data_frames["d1291"])
df = remove_rank(df)
create_scaling_speedup_plot(df, "d1291", xticks_n=xticks_n, scaling_thresholds=scaling_thresholds,
                            static_ylim=True, y_scaling_limits_speedup=y_scaling_limits_speedup, yticks_speedup=yticks_speedup)
```
